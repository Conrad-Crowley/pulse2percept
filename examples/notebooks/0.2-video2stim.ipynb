```python
import numpy as np
import pickle
import pulse2percept as p2p
```

```python
# Place an Argus I array on the retina
argus = p2p.implants.ArgusII(x_center=0, y_center=0, h=50, rot=0*np.pi/180)
```

```python
videopath = '../../data/'
videofiles = [
    'boston-train.mp4',
    'kid-pool.avi'
]
```

```python
for file in videofiles:
    for enc in ['amplitude', 'frequency']:
        fname = videopath + file
        stimfile = 'stim-%s-%s.dat' % (file[:-4], enc)
        try:
            pickle.load(open(stimfile, "rb"))
            print('File %s already exists, skip' % stimfile)
        except:
            print('Processing file %s' % stimfile)
            stim = p2p.stimuli.video2pulsetrain(fname, argus,
                                    framerate=30, coding=enc,
                                    valrange=[0, 60], max_contrast=False,
                                    rftype='square', invert=False)
            pickle.dump(stim, open(stimfile, "wb"))
            print('Created file %s' % stimfile)
```
